```python
# For Data Manupulation
import numpy as np 
import pandas as pd
# For Data Visualisation 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
# For Ignoreing warnings
import warnings
warnings.filterwarnings('ignore')

#for stationarity checking 
from statsmodels.tsa.stattools import adfuller, kpss
#for q,d,p s value 
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
#for training model sarima 
from statsmodels.tsa.statespace.sarimax import SARIMAX
from xgboost import XGBRegressor
#to evaluation of model
from sklearn.metrics import mean_squared_error, mean_absolute_error
#to data spliting 
from sklearn.model_selection import train_test_split,GridSearchCV,TimeSeriesSplit
```

```python
#loading DataSets
df=pd.read_csv(r"training_data.csv")
```

```python
df.head(1)
```

```python
#week data to date type
df['week']=pd.to_datetime(df['week'])
```

```python
#create new variable
df1=df.drop(columns=['record_ID']).sort_values(['week','store_id','sku_id']).copy()
#set week data in columns in properly week form
df1['week'] = df1['week'] + pd.offsets.Week(weekday=6)


#creating new features related to date variable
df1['weekofyear'] = df1['week'].dt.isocalendar().week.astype(int)
df1['month'] = df1['week'].dt.month
df1['quarter'] = df1['week'].dt.quarter
df1['year'] = df1['week'].dt.year
#create new columns that show discount or not 
df1["price_ratio"] = df1["total_price"] / df1["base_price"]


lags = [1,2,3,4,8,12]

for lag in lags:
    df1[f'lag_{lag}'] = (
        df1.groupby(['store_id','sku_id'])['units_sold']
        .shift(lag)
    )

windows = [4,8,12]

#shift1-This prevents data leakage (model seeing future sales).
for w in windows:
    df1[f'rolling_mean_{w}'] = (
        df1.groupby(['store_id','sku_id'])['units_sold']
        .shift(1)      
        .rolling(w)
        .mean()
    )



df1.set_index('week',inplace=True)
```

```python
df1.head(2)
```

```python
#for trail diving data into train and test select 
train = df1[df1['week'] < '2013-06-01']
test  = df1[df1['week'] >= '2013-06-01']

#diving data x and y train and test
X_train = train.drop(columns=['units_sold','week'])
y_train = train['units_sold']

X_test = test.drop(columns=['units_sold','week'])
y_test = test['units_sold']

#create Xgboost model 

model = XGBRegressor(
    n_estimators=500,
    learning_rate=0.01,
    max_depth=4,
    min_child_weight=5,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.5,
    reg_lambda=1.5,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
#Prepare Data for Forecasting (Next 12 Weeks)

last_data = df1.sort_values('week').groupby(['store_id','sku_id']).tail(1)
print(last_data.shape)
last_data.head()


# Copy last known data as starting point
forecast_df = last_data.copy()

# Create an empty dataframe to store all future forecasts
all_forecasts = pd.DataFrame()

# Number of weeks to forecast
n_weeks = 12

for i in range(n_weeks):
    # Step 1: Move week forward
    forecast_df['week'] = forecast_df['week'] + pd.Timedelta(weeks=1)

    # Step 2: Update date-related features
    forecast_df['weekofyear'] = forecast_df['week'].dt.isocalendar().week.astype(int)
    forecast_df['month'] = forecast_df['week'].dt.month
    forecast_df['quarter'] = forecast_df['week'].dt.quarter
    forecast_df['year'] = forecast_df['week'].dt.year

    # Step 3: Prepare model input
    X_next = forecast_df.drop(columns=['week','predicted_units_sold','units_sold'], errors='ignore')

    # Step 4: Predict units_sold
    forecast_df['predicted_units_sold'] = model.predict(X_next)

    # Step 5: Clip negatives
    forecast_df['predicted_units_sold'] = forecast_df['predicted_units_sold'].clip(lower=0)

     # Step 6: Store predictions for this week
    all_forecasts = pd.concat([all_forecasts, 
                               forecast_df[['store_id','sku_id','week','predicted_units_sold']]], 
                              ignore_index=True)

    # Step 7: Update lag features for next iteration
    forecast_df['lag_12'] = forecast_df['lag_8']
    forecast_df['lag_8'] = forecast_df['lag_4']
    forecast_df['lag_4'] = forecast_df['lag_3']
    forecast_df['lag_3'] = forecast_df['lag_2']
    forecast_df['lag_2'] = forecast_df['lag_1']
    forecast_df['lag_1'] = forecast_df['predicted_units_sold']

    # Step 8: Update rolling means
    forecast_df['rolling_mean_4'] = forecast_df[['lag_1','lag_2','lag_3','lag_4']].mean(axis=1)
    forecast_df['rolling_mean_8'] = forecast_df[['lag_1','lag_2','lag_3','lag_4','lag_8']].mean(axis=1)
    forecast_df['rolling_mean_12'] = forecast_df[['lag_1','lag_2','lag_3','lag_4','lag_8','lag_12']].mean(axis=1)


# Reset index
all_forecasts.reset_index(drop=True, inplace=True)

# Show first few rows of 12-week forecast
all_forecasts.head(20)

#checking for store_id 9221 and sku_id 219029

next_12_week_forcasting=all_forecasts[(all_forecasts['store_id']==9221) & (all_forecasts['sku_id']==219029)].reset_index(drop=True)
next_12_week_forcasting.index+=1
next_12_week_forcasting
```
